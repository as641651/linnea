## Noise Analysis

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams.update({'font.size': 14})
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 

def plot_noise(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    plt.rcParams["figure.figsize"] = (20,3)
    plt.yscale('log')
    plt.xlabel("Time (s)")
    plt.ylabel("t - t_min (log scale)")
    plt.plot(x,y,'.')
    plt.show()

    
def plot_noise_r(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    plt.rcParams["figure.figsize"] = (20,3)
    plt.plot(x,y,'.')
    plt.show()
```

```python
def plot_hist(log,l=1e-7,u=1e-2,b=1000):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    #plt.xscale('log')
    plt.xlabel("t - t_min")
    plt.hist(y,bins=b, range=(l,u),color="red")
    plt.show()
```

```python
def plot_quantile(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("GF/s")
    plt.plot(x,y,'.')
    plt.show()
```

```python
plt.title("DGEMM")
plt.rcParams.update({'font.size': 14})
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
plt.rcParams["figure.figsize"] = (7,5)
plot_quantile("quantile.txt")
```

```python
def plot_hist_quantile(log,size,l=1e-7,u=1e-2,b=1000):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            if int(line.split()[0]) == size:
                x.append(float(line.split()[0]))
                y.append(float(line.split()[1]))

    #plt.xscale('log')
    plt.xlabel("GFLOPs/Sec")
    plt.hist(y,bins=b, range=(l,u),color="red")
    plt.show()
```

```python
plot_hist_quantile("quantile_perf.txt",2000,l=0,u=200,b=50)
```
