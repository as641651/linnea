```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.rcParams.update({'font.size': 14})
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
plt.rcParams["figure.figsize"] = (30,10)

def plot_quantile_compare_perf1(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            m = float(line.split()[0])
            n = float(line.split()[1])
            k = float(line.split()[2])
            time = float(line.split()[3])
            flops = float(line.split()[4])
            
            flop_bytes = flops/(1.1*m*n + 1.2*n*k + 1.3*m*k)
            x.append(flop_bytes)
            #x.append(m) 
            y.append((flops/time)*1e-9)
            
    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("F/B")
    plt.ylabel("GF/s")
    plt.plot(x,y,'.',color='b')
    plt.show()
```

```python
#plt.xscale('log')
plt.xlim(300,302)
plot_quantile_compare_perf1("advanced_tests/logs/gemm_mnk.txt")
```
