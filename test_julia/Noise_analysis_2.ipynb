```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams.update({'font.size': 14})
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 

def plot_noise(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    plt.rcParams["figure.figsize"] = (20,3)
    plt.yscale('log')
    plt.xlabel("Time (s)")
    plt.ylabel("t - t_min (log scale)")
    plt.plot(x,y,'.')
    plt.show()

    
def plot_noise_r(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    plt.rcParams["figure.figsize"] = (20,3)
    plt.plot(x,y,'.')
    plt.show()
    
def plot_hist(log,l=1e-7,u=1e-2,b=1000):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    #plt.xscale('log')
    plt.xlabel("t - t_min")
    plt.hist(y,bins=b, range=(l,u),color="red")
    plt.show()
    
def log_hist(log,l=-7,u=-2,b=100):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    bins = np.logspace(l,u,b)
    plt.xscale('log')
    plt.xlabel("t - t_min (log scale)")
    plt.hist(y,bins=bins)
    plt.show()
    
    
def log_hist_proper(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(np.log(float(line.split()[1])))

    plt.hist(y,bins=100,range=(np.log(1e-7), np.log(1e-2)))
    plt.show()
```

```python
def plot_quantile(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("t(s)")
    plt.plot(x,y,'.')
    plt.show()
    
def plot_quantile_perf(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            tmp_x = float(line.split()[0])
            tmp_y = float(line.split()[1])
            tmp_y = ((2*tmp_x*tmp_x*tmp_x)/tmp_y)*1e-9
            x.append(tmp_x)
            y.append(tmp_y)
            

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("GF/s")
    plt.plot(x,y,'.')
    plt.show()
    
def plot_quantile_perf_matvec(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            tmp_x = float(line.split()[0])
            tmp_y = float(line.split()[1])
            tmp_y = ((2*tmp_x*tmp_x)/tmp_y)*1e-9
            x.append(tmp_x)
            y.append(tmp_y)
            

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("GF/s")
    plt.plot(x,y,'.')
    plt.show()
```

```python
def plot_hist_quantile(log,size,l=0,u=20,b=1000):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            if int(line.split()[0]) == size:
                tmp_x = float(line.split()[0])
                tmp_y = float(line.split()[1])
                tmp_y = ((2*tmp_x*tmp_x)/tmp_y)*1e-9
                x.append(tmp_x)
                y.append(tmp_y)

    #plt.xscale('log')
    plt.xlabel("GFLOPs/Sec")
    plt.hist(y,bins=b, range=(l,u),color="red")
    plt.show()
```

```python
def plot_quantile_compare(log1,log2):
    with open(log1) as f:
        lines = f.readlines()
        x = []
        y1 = []
        for line in lines:
            x.append(float(line.split()[0]))
            y1.append(float(line.split()[1]))
            
    with open(log2) as f:
        lines = f.readlines()
        y2 = []
        for line in lines:
            y2.append(float(line.split()[1]))

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("log t(s)")
    plt.plot(x,y1,'.',color='b')
    plt.plot(x,y2,'.',color='r')
    plt.show()
```

```python
def plot_quantile_compare_perf(log1,log2):
    with open(log1) as f:
        lines = f.readlines()
        x = []
        y1 = []
        for line in lines:
            x.append(float(line.split()[0]))
            time = float(line.split()[1])
            cost = float(line.split()[2])
            y1.append((cost/time)*1e-9)
            
    with open(log2) as f:
        lines = f.readlines()
        y2 = []
        for line in lines:
            time = float(line.split()[1])
            cost = float(line.split()[2])
            y2.append((cost/time)*1e-9)

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("GF/s")
    plt.plot(x,y1,'.',color='b')
    plt.plot(x,y2,'.',color='r')
    plt.show()
```

### Matrix Chain (ABC) - paranthesis comparision - (AB)C vs A(BC)

> $A,B \in \mathbb{R}^{n \times n} \\ C \in \mathbb{R}^{n \times n/2}$

**Algorithm - Blue** : 

> (AB)C ; cost = $2n^{3} + 2n^{2}\frac{n}{2} = 3n^{3}$

**Algorithm - Red** : 

> A(BC) ; cost = $2n^{2}\frac{n}{2} + 2n^{2}\frac{n}{2} = 2n^{3}$

```python
plt.yscale('log')
plot_quantile_compare("log_noise/c1.txt","log_noise/c2.txt")
```

```python
plot_quantile_compare_perf("log_noise/c1.txt","log_noise/c2.txt")
```

```python
def plot_quantile_compare_fb(log1,log2):
    
    with open(log1) as f:
        lines = f.readlines()
        x1 = []
        y1 = []
        for line in lines:
            s = float(line.split()[0])
            mem_trans = 2*s*s + s*s
            
            time = float(line.split()[1])
            cost = float(line.split()[2])
            fb = cost/mem_trans
            
            x1.append(fb)
            y1.append((cost/time)*1e-9)
            #y1.append(time)
            
    with open(log2) as f:
        lines = f.readlines()
        x2 = []
        y2 = []
        for line in lines:
            s = float(line.split()[0])
            mem_trans = 2*s*s + s*s
            
            time = float(line.split()[1])
            cost = float(line.split()[2])
            fb = cost/mem_trans
            
            x2.append(fb)
            y2.append((cost/time)*1e-9)
            #y2.append(time)

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("F/B")
    plt.ylabel("GF/s")
    plt.plot(x1,y1,'.',color='b')
    plt.plot(x2,y2,'.',color='r')
    plt.show()
```

```python
plot_quantile_compare_fb("log_noise/c1.txt","log_noise/c2.txt")
```

```python
plot_quantile_compare_fb("log_noise/c1.txt","log_noise/c2.txt")
```
