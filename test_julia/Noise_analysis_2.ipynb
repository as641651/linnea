```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams.update({'font.size': 14})
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 

def plot_noise(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    plt.rcParams["figure.figsize"] = (20,3)
    plt.yscale('log')
    plt.xlabel("Time (s)")
    plt.ylabel("t - t_min (log scale)")
    plt.plot(x,y,'.')
    plt.show()

    
def plot_noise_r(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    plt.rcParams["figure.figsize"] = (20,3)
    plt.plot(x,y,'.')
    plt.show()
    
def plot_hist(log,l=1e-7,u=1e-2,b=1000):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    #plt.xscale('log')
    plt.xlabel("t - t_min")
    plt.hist(y,bins=b, range=(l,u),color="red")
    plt.show()
    
def log_hist(log,l=-7,u=-2,b=100):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    bins = np.logspace(l,u,b)
    plt.xscale('log')
    plt.xlabel("t - t_min (log scale)")
    plt.hist(y,bins=bins)
    plt.show()
    
    
def log_hist_proper(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(np.log(float(line.split()[1])))

    plt.hist(y,bins=100,range=(np.log(1e-7), np.log(1e-2)))
    plt.show()
```

```python
def plot_quantile(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            x.append(float(line.split()[0]))
            y.append(float(line.split()[1]))

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("t(s)")
    plt.plot(x,y,'.')
    plt.show()
    
def plot_quantile_perf(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            tmp_x = float(line.split()[0])
            tmp_y = float(line.split()[1])
            tmp_y = ((2*tmp_x*tmp_x*tmp_x)/tmp_y)*1e-9
            x.append(tmp_x)
            y.append(tmp_y)
            

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("GF/s")
    plt.plot(x,y,'.')
    plt.show()
    
def plot_quantile_perf_matvec(log):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            tmp_x = float(line.split()[0])
            tmp_y = float(line.split()[1])
            tmp_y = ((2*tmp_x*tmp_x)/tmp_y)*1e-9
            x.append(tmp_x)
            y.append(tmp_y)
            

    #plt.yscale('log')
    #plt.xscale('log')
    plt.xlabel("size")
    plt.ylabel("GF/s")
    plt.plot(x,y,'.')
    plt.show()
```

```python
def plot_hist_quantile(log,size,l=0,u=20,b=1000):
    with open(log) as f:
        lines = f.readlines()
        x = []
        y = []
        for line in lines:
            if int(line.split()[0]) == size:
                tmp_x = float(line.split()[0])
                tmp_y = float(line.split()[1])
                tmp_y = ((2*tmp_x*tmp_x)/tmp_y)*1e-9
                x.append(tmp_x)
                y.append(tmp_y)

    #plt.xscale('log')
    plt.xlabel("GFLOPs/Sec")
    plt.hist(y,bins=b, range=(l,u),color="red")
    plt.show()
```
