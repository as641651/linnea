```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.rcParams.update({'font.size': 14})
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 
plt.rcParams["figure.figsize"] = (30,10)
```

### Matchain mnkl

```python
log = "advanced_tests/logs/matchain_mnkl.txt"
with open(log) as f:
    lines = f.readlines()
    x = []
    y_time = []
    y_perf = []
    for line in lines:
        m = float(line.split()[0])
        n = float(line.split()[1])
        k = float(line.split()[2])
        l = float(line.split()[3])
        time = float(line.split()[4])
        flops = float(line.split()[5])

        mem_ref = m*n + n*k + m*k + m*k + k*l + m*l 
        flop_bytes = flops/mem_ref
        x.append(flop_bytes)
        #x.append(m) 
        y_time.append(time)
        y_perf.append((flops/time)*1e-9)
            
```

```python
#plt.yscale('log')
#plt.xscale('log')
#plt.xlim(300,302)
plt.xlabel("F/B")
plt.ylabel("GF/s")
plt.plot(x,y_perf,'.',color='b')
plt.show()
```

```python
#plt.yscale('log')
#plt.xscale('log')
#plt.xlim(300,350)
plt.xlabel("F/B")
plt.ylabel("t")
plt.plot(x,y_time,'.',color='b')
plt.show()
```

### Matchain nn
